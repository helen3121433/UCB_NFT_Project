```python
# df_list = ['magicticket','guild_saga_heroes','monkelabs_dao','galactic_geckos','magnum_ai','degenerate_ape_academy']
# magicticket = 'magicticket'
almost_famous_pandas = 'almost_famous_pandas'

def split_attribute(dfname):
        #pull out only data that match the collecion name
        name_df = all_df[all_df['collectionName'] == dfname]
        # split trait_type and trait_value into separate columns
        split_att_df = name_df['attributes'].str.split(',', expand=True)
        # print the list of row 1
        col_name_list = split_att_df.iloc[0]
        # only keep the value columns in attribute_df
        split_att_df = split_att_df.loc[:, np.arange(len(split_att_df.columns)) %2 != 0]
        # use regular expression to clean ('value') and non word characters
        split_att_df = split_att_df.replace({'(\'value\')?(\W)':''}, regex=True)
        # return list, collection_df, and attribute_df
        return col_name_list,name_df, split_att_df


 #list[0] = column name
#list[1] = name_df
#list[2] = attribute_df       
amf_df = split_attribute(almost_famous_pandas)
amf_df


```

```python
name_list = ['Trait_Outfit','Trait_Head','Trait_Glasses','Trait_Eyes','Trait_Extras','Trait_Skin','Trait_Mouth','Trait_Body','Trait_Background','Trait_Face']

def add_att(trait_list,name_df,att_df):
    #list of column name of current attribute_df
    col_list = list(att_df)
    # change the column name of attribute_df 
    for x in range(len(col_list)):
        att_df.rename(columns={col_list[x]:trait_list[x]}, inplace=True)
    
    # Drop orginal attribute columns of collection_df
    name_df = name_df.drop(columns=['attributes'])
    
    #use index as join key to join collection_df and attribute_df
    name_df = name_df.merge(att_df, how='outer',left_index=True,right_index=True)
    
    #Reset index
    name_df = name_df.reset_index()
    name_df = name_df.drop(columns='index')
    return name_df

# call functions.
new_amf_pd = add_att(name_list,amf_df[1],amf_df[2])

```
