```python
import pandas as pd

v1_solana_df = pd.read_csv("solana_nft_full/solana_v1/Solana_NFT_Collections_v1/collections.csv", index_col=[0])
v2_solana_df = pd.read_csv("solana_nft_full/solana_v2/Solana_NFT_Collections_v2/collections.csv", index_col=[0])
v3_solana_df = pd.read_csv("solana_nft_full/solana_v3/Solana_NFT_Collections_v3/collections.csv", index_col=[0])
v4_solana_df = pd.read_csv("solana_nft_full/solana_v4/Solana_NFT_Collections_v4/collections.csv", index_col=[0])
```

```python
c_frame = [v1_update_df,v2_update_df,v3_update_df,v4_update_df]
collection_list_concat = pd.concat(c_frame, join='inner', ignore_index=True)

print(collection_list_concat.shape)
collection_list_concat.head()
```

```python
new_collection_list = collection_list_concat
new_collection_list
```

```python
new_collection_list = new_collection_list.sort_values(by=['symbol'])
new_collection_list
```

```python
new_collection_list['symbol'].value_counts()
```

```python
appear_number_list = new_collection_list['symbol'].value_counts()
appear_number_list_df = pd.DataFrame(appear_number_list)
appear_number_list_df.reset_index(inplace=True)
appear_number_list_df = appear_number_list_df.rename(columns={"index": "symbol", "symbol": "#_appear"})
appear_number_list_df
```

```python
keep_list_df = appear_number_list_df[appear_number_list_df['#_appear'] == 4]
keep_list_df
```
